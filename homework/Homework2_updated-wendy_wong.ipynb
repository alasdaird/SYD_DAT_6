* Imagine you are trying to explain to someone what Linear Regression is - but they have no programming/maths experience? How would you explain the overall process, what R-Squared means and how to interpret the coefficients?

 *I would start by explaining to someone that Linear Regression is a modelling technique that allows you to select  a minuimum of one or more parameters or independent variables to include in the model to predict the dependent variable or outcome variable which is your target.  This is called a Supervised Learning technique as you  can specifically choose your variables and it is not a random selection process. To assist with  naming your dependent or predictor variables , you will have to think about the potential independent variable(s)  to include in the model which will affect your outcome or response variable that you are trying to predict. An example for an insurance company could be that the company want to predict how television advertising will affect sales revenue, therefore in a  simple linear model the dependent variable (Y) will be Sales Revenue which is the outcome we are trying to predict based on exposure to television advertising (X) which is the indepedent variable. Hence a rule of thumb is to think in terms of Y depends on Xi (i.e. Sales revenue depends on television advertising, customer service and management). 
 
We use a method called Ordinary Least Squares (OLS) to try to  fit a linear model (linear line) across all observed values and try to minimise the distance between the residual sum of squares (error term) and observations within the data points.

The data needs to be visualised to check the actual fit of the model, under a linear regression model, you want to ensure that the data points are not too overfitted but will be fitted  to represent a linear relationship under the assumption that the mean of the error is 0 and the data points are normally distributed.

The output of the linear regression model produces R-squared which measures the accuracy of the model or how well are the data points explained by variance in the model. R-squared differs by industry for example social sciences have a lower R-quared of 30% which is acceptable, however in general the higher R-squared value such as 80% is an indication of a good model because 80% of the variance is explained by the model.

To intepret the coefficients of a simple linear regression model  with only one feature variable, it is represented by the  linear regression model:

Y = B0+ B1X1 + E

Sales_Revenue = B0 + Television_Advertising x B1 + Error

Sales_Revenue = 20
Television_Advertising = 0.15

Hence the output of the regression coefficients can be interpreted that for a one unit change in television advertising, the sales revenue will increase by 0.15 units after controlling for all the other variables within the model.
