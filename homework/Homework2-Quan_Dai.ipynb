### Python Programming Skills - Titanic Data Set

```python
###Charting exercise

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

titanic = pd.read_csv('titanic.csv',index_col='PassengerId')
titanic.dropna(inplace=True)
titanic.shape
#histogram using pandas
titanic.hist(column='Age')
plt.show()
#histogram using pyplot
plt.hist(titanic.Age)
plt.show()
#histogram using seaborn
sns.distplot(titanic.Age)
plt.show()
```

```python
### Grouping Data

#3rd class male survived
survivedcount = len(titanic[(titanic.Pclass ==3) & (titanic.Survived ==1) & (titanic.Sex =='male')])
notsurvivedcount = len(titanic[(titanic.Pclass ==3) & (titanic.Survived ==0) & (titanic.Sex =='male')])
print ' 3rd class male passengers survived =',survivedcount
print ' 3rd class male passengers not survived =',notsurvivedcount

#1st class female survived
survivedcount = len(titanic[(titanic.Pclass ==1) & (titanic.Survived ==1) & (titanic.Sex =='female')])
notsurvivedcount = len(titanic[(titanic.Pclass ==1) & (titanic.Survived ==0) & (titanic.Sex =='female')])
print ' 1st class female passengers survived =',survivedcount
print ' 1st class female passengers not survived =',notsurvivedcount

# group by data
titanic['Headcount'] = 1
groupbyclass = titanic.groupby(titanic.Pclass).sum()
groupbyclass['Survivepercent'] = groupbyclass.Survived/groupbyclass.Headcount
groupbyclass['Avgfare'] = groupbyclass.Fare/groupbyclass.Headcount
groupbyclass
```

### Explain linear regression

Linear regression is a statistical tool we have which can be used to predict a continous value. In involves having a set of features for which we assume have a linear relationship with the predictor which we are trying to estimate.

For each of our features we need to calculate the coefficient which translates directly as for each unit change in the feature results in a change of the coefficient on the predictor we are trying to estimate.

R squared is a measure of how well our regression line fits our data points. It can be interpreted also as the amount of explained variation vs the total variation in the model. If our model has 1 R-squared then all the variation in the model is explained by our model.
