## Communication

Linear regression is an analysis technique that looks to describe the relationship between at least two variables. At least one of these variables is dependent on the other, the degree to which is what regression analysis aims to determine. 

The process of linear regression typically involves plotting data points on a scatter plot and visually determining if any clear relationship exists between the two variables. Once determined, the aim is then to plot a line (model) through these points that is the closest fit to the data. The measure of the closeness of fit of this line to the data is referred to as R-squared and sits between 0%-100%.
  - R-squared = 0% - indicates that the model explains none of the variability of the response data around its mean
  - R-squared = 100% - indicated that the model explains all of the variability of response data around its mean
